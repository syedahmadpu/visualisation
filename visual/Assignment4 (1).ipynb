```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
br_state = pd.read_excel('Bihar_state.xls')
ind_data = pd.read_excel('DDWC-000009.xls')
#up_state = pd.read_excel('DDWC-090009.xls')
br_state = br_state.where(br_state['Area Name']=='State - BIHAR  10').dropna()
br_state = br_state.where(br_state['Age-group']=='Total').dropna()
br_state = br_state.where(br_state['Total/ Rural/ Urban']=='Total').dropna()
br_state = br_state.where(br_state['Total/ Rural/ Urban']=='Total').dropna()
ind_data = ind_data.where(ind_data['Total/ Rural/ Urban']=='Total').dropna()
ind_data = ind_data.where(ind_data['Area Name']=='INDIA').dropna()
ind_data = ind_data.where(ind_data['Age-group']=='Total').dropna()
#up_state = up_state.where(up_state['Area Name']=='State - UTTAR PRADESH 09').dropna()
#up_state = up_state.where(up_state['Age-group']=='Total').dropna()
#up_state = up_state.where(up_state['Total/ Rural/ Urban']=='Total').dropna()
br_state['State Literacy Rate'] = br_state['Literate - Persons']/br_state['Total population - Persons']
ind_data['Country Literacy Rate'] = ind_data['Literate - Persons']/ind_data['Total population - Persons']
#up_state['state Literacy Rate'] = up_state['Literate - Persons']/up_state['Total population - Persons']
cols = ['State Literacy Rate','Religion']
br_state=br_state[cols]
br_state['Country Literacy Rate']=ind_data['Country Literacy Rate']
#br_state['UP Literacy Rate']=up_state['state Literacy Rate']
plt.style.use('fivethirtyeight')

br_state[['State Literacy Rate','Country Literacy Rate']].plot(br_state['Religion'],kind = 'bar',
                            width =0.4, figsize=(6,6),alpha=0.8,
                            legend=True, fontsize=12)
plt.subplots_adjust(bottom=0.2)
ax=plt.gca()
ax.set_ylabel('Literacy Rate')
ax.set_title('Literacy Rate of Religious communities of Bihar(state) vs Literacy Rate of Religious communities of India ',size = 12)
for tics in ax.xaxis.get_ticklabels():
    tics.set_rotation(45)
tick_labels = [item.get_text() for item in ax.xaxis.get_ticklabels()]
tick_labels[0]='All'
tick_labels[7]='Others'
ax.xaxis.set_ticklabels(tick_labels)
legend_labels=['','']
legend_labels[0] = 'Literacy Rate of Bihar state'
legend_labels[1] = 'Literacy Rate of India'
ax.legend(legend_labels , loc = 'best')#loc = best gives best place for legends
#ind_data.head()
```
